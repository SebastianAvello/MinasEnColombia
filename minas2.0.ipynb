```python
#Fuente Datos: 
#https://www.datos.gov.co/Inclusi-n-Social-y-Reconciliaci-n/Situaci-n-V-ctimas-Minas-Antipersonal-en-Colombia/yhxn-eqqw


#Fuente: Agregar imagenes a Dash
#https://dash.plotly.com/external-resources

import pandas as pd

import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px #(need to pip install plotly==4.4.1)

df = pd.read_csv("dataminas.csv")

# you need to include __name__ in your Dash constructor if
# you plan to use a custom CSS or JavaScript in your Dash apps
app = dash.Dash(__name__)

#---------------------------------------------------------------
app.layout = html.Div([
    
    html.Div([
         html.Img(src='/assets/image.jpg'),     
    dcc.Dropdown(id='my_dropdown',
            options=[
                     {'label': 'Departamento', 'value': 'departamento'},
                     {'label': 'Año', 'value': 'ano'},
                     {'label': 'Tipo de Área', 'value': 'tipoarea'},
                     {'label': 'Género', 'value': 'genero'},
                     {'label': 'Estado del Afectado', 'value': 'estado'},
                     {'label': 'Condicion', 'value': 'condicion'},
                     {'label': 'Actividad de la Afectación', 'value': 'Actividad'}

           ],
            value='ano',
            multi=False,
            clearable=False,
            style={"width": "50%"}
        ),
    ]),

    html.Div([
        dcc.Graph(id='the_graph')
    ]),

])

#---------------------------------------------------------------
@app.callback(
    Output(component_id='the_graph', component_property='figure'),
    [Input(component_id='my_dropdown', component_property='value')]
)

def update_graph(my_dropdown):
    dff = df

    piechart=px.pie(
            data_frame=dff,
            names=my_dropdown,
            hole=.3,
            )

    return (piechart)


if __name__ == '__main__':
    app.run_server(debug=False)
```
